# Extract
This notebook extracts text from the documents in raw_data to create datasets for each language

```python
import re
from pathlib import Path
```

```python
PATH_TO_RAW_DATA = Path("../data/raw_data")
PATH_TO_PROCESSED_DATA = Path("../data/processed_data")
ALLOWED_CHARACTERS = "[^a-zA-ZáéíóúÁÉÍÓÚüÜñÑàèùôêâûîôÂÊÎÔÛçÇëËïÏÿŸ ]"
MIN_WORDS_IN_SENTENCE = 5
```

```python

def get_list_of_sentences_from_text(complete_text: str, min_sentence_length = MIN_WORDS_IN_SENTENCE):
    lines = [' '.join(line.strip().split()) for line in complete_text.split("\n")]
    lines = [re.sub(ALLOWED_CHARACTERS, "", line) for line in lines]
    return [line for line in lines if len(line.split()) > MIN_WORDS_IN_SENTENCE]


for language_path in PATH_TO_RAW_DATA.iterdir():
    language = language_path.stem
    all_text_for_language = []
    for text_file in language_path.iterdir():
        all_text_for_language += get_list_of_sentences_from_text(text_file.read_text())
    print(f"{language=}. Number of lines of text {len(all_text_for_language)}")
    all_text_for_language_str = "\n".join(all_text_for_language)
    (PATH_TO_PROCESSED_DATA / f"{language}.txt").write_text(all_text_for_language_str)

all_text
```
